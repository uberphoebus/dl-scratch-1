```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.array(x > 0, dtype=np.int32)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.title("Step Function")
plt.ylim(-0.1, 1.1)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.title("ReLU Function")
plt.xlim(-6, 6)
plt.show()
```
